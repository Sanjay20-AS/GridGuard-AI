```python
from __future__ import annotations

import sys
from pathlib import Path

import torch


def find_repo_root() -> Path:
    """Find the repository root so the notebook can import src modules reliably."""
    candidate_roots = [
        Path.cwd(),
        Path.cwd() / "GridGuard-AI",
        Path("/content/GridGuard-AI"),
        Path("/content/drive/MyDrive/GridGuard-AI"),
        Path("/content/drive/MyDrive/gridguard-ai"),
    ]

    for candidate_root in candidate_roots:
        for search_root in (candidate_root, *candidate_root.parents):
            if (search_root / "src" / "model.py").exists():
                return search_root

    raise FileNotFoundError(
        "Could not locate the repository root containing src/model.py. "
        "In Colab, clone or mount the repo first, then set REPO_ROOT manually if needed."
    )


repo_root = find_repo_root()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from src import model as model_module

if not torch.cuda.is_available():
    raise RuntimeError(
        "Enable a GPU in Colab first: Runtime -> Change runtime type -> GPU."
    )

model_module.set_random_seed()
model_module.get_device = lambda: torch.device("cuda")
print(f"Training on: {torch.cuda.get_device_name(0)}")
model_module.main()
```
